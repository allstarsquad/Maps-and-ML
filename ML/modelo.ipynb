```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
df = pd.read_excel('public_imoveis_spiunet_150122.xlsx')
```

Drop de colunas que não são importantes/interessantes para realizar a previsão.

```python
df.drop(['Rip Imóvel', 'Nº/RIP Utilização', 'Endereço', 'Matrícula', 'Nome do Cartório', 
'Cód. UG/CPF/CNPJ do Resp.', 'Proprietário Oficial', 'Responsável/UG', 'Data do Cadastro do Imóvel', 'Regime de Utilização', 
'Longitude', 'Latitude', 'Valor Contabilizado (R$)'], axis=1, inplace=True)
```

```python
df.head()
```

Processamento da Data de Registro.

```python
df['Data de Registro no Cartório'] = pd.to_datetime(df['Data de Registro no Cartório'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
```

```python
df['Data_Ano'] = df['Data de Registro no Cartório'].dt.year
df.drop(['Data de Registro no Cartório'], axis=1, inplace=True)
```

```python
df.head()
```

Drop de linhas que possuam valores nulos.

```python
df = df.dropna()
```

```python
df.info()
```

Achamos melhor realizar um modelo com uma tripla filtragem (UF -> Município -> Bairro), pois dados categóricos, como as UF's, os Municípios, Bairros e CEP's são bem complicados de serem processados pelo modelo preditivo, existem algumas alternativas, como realizar um [One Hot Encoding](https://www.educative.io/blog/one-hot-encoding), ou um [Ordinal Encoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OrdinalEncoder.html), porém, um One Hot Encoding, para pelos menos todos os bairros de certo município e os CEP's, iria gerar um dataset com MUITAS colunas (mais de 1000), e isso é muito ruim para o modelo analisar, e além disso, o Ordinal Encoder também é bem ruim, quando existem muitos valores, então fazer essa filtragem foi a melhor alternativa.

```python
df['Bairro'].value_counts()
```

Identifico a ASA NORTE como o bairro que possui mais imóveis (quanto mais imóveis, mais o modelo será preciso). 

```python
df[df['Bairro']=='ASA NORTE'].head()
```

```python
df_asanorte = df[(df['UF'] == 'DF') & (df['Município'] == 'Brasilia') & (df['Bairro'] == 'ASA NORTE')]
```

```python
df_asanorte.head()
```

```python
df_asanorte.drop(['UF', 'Município', 'Bairro'], axis=1, inplace=True)
```

```python
df_asanorte.head()
```

Passo os CEP's e o Tipo de Imóvel para o One Hot Encoding

```python
df_asanorte = pd.get_dummies(df_asanorte, columns=['CEP', 'Tipo de Imóvel'])
```

Como a Natureza é binária, só precisa transformar para 0 e 1

```python
natureza = {'Urbano':0, 'Rural':1}
df_asanorte['Natureza do Terreno'] = df_asanorte['Natureza do Terreno'].map(natureza)
```

```python
df_asanorte
```

```python
df_asanorte.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.tree import ExtraTreeRegressor

model = ExtraTreeRegressor(random_state=0)

X = df_asanorte.drop(['Valor da Utilização (R$)'], axis=1)
y = df_asanorte['Valor da Utilização (R$)']

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.1, random_state=0)

model.fit(X_train, Y_train)

result = model.predict(X_test)
```

```python
diff_rel = 0

for i in range(0, Y_test.size-1):
    val1 = Y_test.iloc[i]
    val2 = result[i]
    #para cada valor, calculo quantos porcento a diferença entre o valor real e o previsto, equivale ao valor real
    diff_rel += abs(val1-val2)/val1*100
    if i<10:
        print("Valor Real: {v1} / Valor Previsto:{v2} / Diferença Relativa:{diff}".format(v1=val1, v2=val2, 
        diff=abs(val1-val2)/val1*100))

#printo a média das diferenças relativas
print("\nDiferença Média Relativa: {res}".format(res=diff_rel/Y_test.size))
```
